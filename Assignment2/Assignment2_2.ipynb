```python
import csv
import os, json
import datetime
# assign the path of the source and destination folders to a variable
path_to_folder='DataProcessed'
path_to_json = 'Data/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')] #looping in the folder "data" where the the json files are placed 
print (json_files) 
```

```python
def ensure_dir(path_to_folder): #declaring a function 
    if not os.path.exists(path_to_folder): #if the folder doesn't exist then create the directory
        os.makedirs(path_to_folder)# make a directory
```

```python
dict2={}
for js in json_files: #check for files in json files
    with open(os.path.join(path_to_json, js)) as json_file: #append the path of file to the source folder 
        loaded_file = json.load(json_file) #load the file to be selected
        for key,value in loaded_file.items(): #for every key,value pair in the dictionary
            if key == 'location': # check if key is location
                country = value.get('country') #get the value  of key country
                state=value.get('state')
                city=value.get('city')
                zipcode=value.get('zip_code')
                path=os.path.join(path_to_folder +'/' + country) #append the folder to the path
                path=os.path.join(path +'/' + state)
                path=os.path.join(path +'/' + city)
                path=os.path.join(path +'/' + zipcode)
                #print(path)  
            if key=='term':
                #print('a')
                term=value
                #print(path) 
                path=os.path.join(path +'/' + term)
                ensure_dir(path)
                os.rename(os.path.join(path_to_json, js) ,os.path.join(path,js))#copy the files from source folder to the destination folder              
```
