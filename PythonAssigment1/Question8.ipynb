```python
import nltk
from nltk.corpus import nps_chat
```

```python
#Words from nps_chat which don't include the word user
file_id=nps_chat.fileids()
word_list=[]
from xml.etree import ElementTree as ET
for id in file_id:
    chat = nps_chat.raw(id)
    my_xml = ET.fromstring(chat)
    for child in my_xml.iter('t'):
        word_list.append(child.get('word')) 
npschat_set=set()
for word in word_list:
    if 'User' not in word:
        npschat_set.add(word.lower());
print("Words from nps_chat which don't include the word user are:",len(npschat_set),"\nwords are:\n","\n",npschat_set)
```
